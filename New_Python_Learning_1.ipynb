# **Assignment, Shallow Copy, Deep Copy**




## **Assignment (=)**

* **Using = assigns a reference to the same object**
* **No actual copying of the object takes place**
* **If the object is mutable (like a list or dictionary), changes through either reference affect the same underlying object**
* **Copies Outer Object? No**
* **Copies Nested Objects? No**
* **Effect on Original When Modifying Copy - Changes affect original and copy**

## **Shallow Copy:**
* **Creates a new object, but inserts references to the objects found in the original**
* **For nested objects, the references to inner objects are copied, not the objects themselves**
* **Changes to nested mutable objects affect both copies**
* **You can create a shallow copy using: The .copy() method (for lists, dicts), The copy module’s copy.copy() function**
* **Copies Outer Object? Yes**
* **Copies Nested Objects? No**
* **Effect on Original When Modifying Copy - Changes to nested objects affect both**
* **.append() adds an item to the outer list, so if you .append() to the shallow-copied list, it modifies the new list only**

## **Deep Copy**
* **Creates a new object and recursively copies all nested objects**
* **The new copy is completely independent**
* **Changes in the copy don’t affect the original, even for nested mutable objects**
* **You can create a deep copy using the copy module’s copy.deepcopy() function**
* **Also use [:]**
* **Copies Outer Object? Yes**
* **Copies Nested Objects? Yes**
* **Effect on Original When Modifying Copy - Changes do not affect original**

## **Notes**
* **No Difference for sets as sets cannot contain mutable types**

- b = a                  # no copy (alias)
- b = a.copy()           # shallow copy (outer only) for built-in containers
- b = copy.copy(a)       # shallow copy (generic)
- b = copy.deepcopy(a)   # deep copy (recursive)


```python
import copy

# Assignment 
a = [1, 2, [3, 4]]
b = a  # b references the same list as a
b[0] = 10
print(f"b references the same list as a{a}, {b}")  # Output: [10, 2, [3, 4]]

# Shallow Copy
a = [1, 2, [3, 4]]
b = copy.copy(a)  # shallow copy
b[0] = 10
b[2][0] = 30
print(f"Changes to nested objects affects both {a}, {b}")  # Output: [1, 2, [30, 4]]


a = [1, 2, [3, 4]]
b = copy.deepcopy(a)  # deep copy
b[2][0] = 30

print(f"Deepcopy does not change original {a}, {b}")  # Output: [1, 2, [3, 4]]

a = [1, 2, [3, 4]]
b = copy.deepcopy(a)  # deep copy
b[2][0] = 30

print(f"Deepcopy does not change original {a}, {b}")  # Output: [1, 2, [3, 4]]

a = [1, 2, [3, 4]]
b = a[:]  # deep copy
b[1] = 300

print(f"Deepcopy does not change original {a}, {b}")  # Output: [1, 2, [3, 4]]
```
