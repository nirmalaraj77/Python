##**Method 1 - figure and subplots**
###**With shortcuts and multiple plots**
* **Create multiple subplots in grid - Small multiples (subplots)**
* **Create figure object with rows and columns of subplots**
* **fig, ax = plt.subplots (r, c)**
* **index into object - use r and c**
* **ax[0,0].plot(x,y)**
###**Special case - two subplots one on top of other**
* **use ax[0], ax[1]**
* **Only 1 xlabel and title**
* **Share y axes range - sharey = True**






```python
# Create multiple Axes with plt.subplots rows and columns
fig, ax = plt.subplots(2,1, sharey=True)

# Plot Seattle data, setting data appearance with shortcuts
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], 'o--b', ms=10)

# Plot Austin data, setting data appearance with shortcuts
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], 'v--r', ms=10)

# Customize the x-axis label
ax[1].set_xlabel("Time (months)")

# Customize the y-axis label
ax[0].set_ylabel ("Precipitation (inches)")
ax[1].set_ylabel ("Precipitation (inches)")

# Add the title
ax[0].set_title ("Weather patterns in Austin and Seattle", loc = 'left')

plt.show()
```

##**twinx Share x axes**
* **ax2 = ax.twinx()**
* **change ticks with ticks_params**

```python
import matplotlib.pyplot as plt

# Initalize a Figure and Axes
fig, ax = plt.subplots()

# Plot the CO2 variable in blue
ax.plot(climate_change.index, climate_change["co2"], color='blue')

# Create a twin Axes that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature in red
ax2.plot(climate_change.index, climate_change["relative_temp"], color='red')

# Change twin axes tick colors with tick_params
ax2.tick_params ('y', colors = 'red')

plt.show()
```

##**Define function**

```python
# Define a function called plot_timeseries
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  # Plot the inputs x,y in the provided color
  axes.plot(x, y, color=color)

  # Set the x-axis label
  axes.set_xlabel(xlabel)

  # Set the y-axis label
  axes.set_ylabel(ylabel, color=color)

  # Set the colors tick params for y-axis
  axes.tick_params('y', colors=color)


# Use the function

fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_change.index, climate_change["co2"], 'blue', "Time (years)", "CO2 levels")

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', "Time (years)", "Relative temperature (Celsius)")

plt.show()
```

##**Annotate**
* **Use ax.annotate (text, xy)**
* **Use pd.Timestamp(x,y)**
* **Place annotate text - xytext =(x,y)**
* **Use arrowprops dict keywords**


```python
fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_change.index, climate_change["co2"], 'blue', "Time (years)", "CO2 levels")

# Create an Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', "Time (years)", "Relative temp (Celsius)")

# Annotate the point with relative temperature >1 degree
ax2.annotate(">1 degree", xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2008-10-06'), -0.2), arrowprops={'arrowstyle':'->', 'color':'gray'})

plt.show()
```

##**Bar Charts - Quantative comparisions**
* **Rotate xlabel 90 degrees set_xticklabels**
* **Stacked bar chart - bottom keyword argument**
* **Set bottom as total of stacks**
* **Add color label - label keyword argument**
* **Add color label - call legend**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
medals = pd.read_csv('https://raw.githubusercontent.com/nirmalaraj77/Python/main/medals.csv', index_col=0)

# Fix error in 'Bronze' column name by renaming
medals = medals.rename(columns={'Bronze ': 'Bronze'})
```

```python
# Create figure and subplots
fig, ax = plt.subplots()

# Bar plot index and gold
ax.bar(medals.index, medals['Gold'], label = 'Gold')

# Add silver
ax.bar(medals.index, medals['Silver'], bottom = medals['Gold'], label = 'Silver')

# Add bronze on top of gold and silver (total of gold + silver)
ax.bar(medals.index, medals['Bronze'], bottom = medals['Gold'] + medals['Silver'], label = 'Bronze')

# Rotate 90 degrees set_xticklabels
ax.set_xticklabels(medals.index, rotation = 90)

# Set ylabel
ax.set_ylabel('Number of medals')

#Call legend
ax.legend()

plt.show()
```

##**Histograms - Quantative comparisions**
* **Distribution of values**
* **Counts distributions in bins**
* **Default binsize = 10**
* **Change binsize - bins keyword argument and integer**
* **Change custom binsize - bins keyword argument and sequence of values**
* **Must use labels and ax.legend**
* **Change histogram type - histtype = 'step'**



```python
import pandas as pd
import matplotlib.pyplot as plt

mens_rowing = pd.read_csv('https://raw.githubusercontent.com/nirmalaraj77/Python/main/rowing.csv', index_col=0)
mens_gymnastics = pd.read_csv('https://raw.githubusercontent.com/nirmalaraj77/Python/main/gymnastics.csv', index_col=0)

# Create figure and axes
fig, ax=plt.subplots()

# Call histograms
ax.hist(mens_rowing['Height'], label='Rowing', bins =[150,160,170,180,190,200,210], histtype='step')
ax.hist(mens_gymnastics['Height'], label = 'Gymnastics', bins =[150,160,170,180,190,200,210], histtype='step')

# Set labels
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')

ax.legend()
plt.show()
```

##**Box plots**
* **shows distribution of quantitative dataacross different groups of a categorical variable**
* **colored box represents 25th to 75th percentile**
* **line in middle represents median**
* **whiskers show spread of distribution - 1.5 times IQR**
* **floating points represent outliers**
* **use kind = 'box'**
* **change order**
* **use sym parameter to customize outliers**
* **empty string to omit outlierrs - sym = ""**
* **use whis parameter to customize whiskers**
* **2 times IQR - whis = 2.0**
* **5th and 95th percentile - whis = [5, 95]**
* **min and max - whis = [0, 100]**
* **use hue to subgroup by color**

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')

sns.catplot(x='day', y='total_bill', data = tips, kind='box', hue = 'time', whis=[5, 95])
```

##**Point plots**
* **show mean of a quantitative variable for the observations in each category**
* **plotted as a single point**
* **vertical bars represent the 95% confidence intervals for that mean**
* **confidence intervals show the level of uncertainty about the mean estimates**
* **Assuming data is a random sample of population - can be 95% sure that true population mean in each group lies within the confidence interval shown**
* **line plot has quantitative variable (time) on x-axis**
* **point plot has catergorical variable on x-axis**
* **use kind = 'point'**
* **segregate with hue**
* **disconnect points = join = False**
* **calculate points and ci from median (instead of mean) - import median from numpy and use estimator = median**
* **median more robust to outliers**
* **use ci=None**
* **use capsize = desired width of cap e.x. 0.2**

##**Changing plot style and color**
###**Figurestyle**
* **change global - sns.set_style('figure_style')**
* **"white" (default) - clean axes with solid white bg**
* **"whitegrid" - add a gray grid in the background**
* **"ticks" - adds small tick marks to the x- and y-axes**
* **"dark"  - gray background**
* **"darkgrid"  - gray background with a white grid**
###**Palette**
* **change global - sns.set_palette('palette_name')**
* **Diverging palette - RdBu, PRGn**
* **add _r to reverse - RdBu_r, PRgn_r**
* **sequential palettes - Greys, Blues, Purples, PuRd, GnBu**
* **default - color_palette()**
* **custom palette - create list with colors or hex codes**
###**Scale**
* **change global - sns.set_context('scale_name')**
* **smallest to largest - paper, notebook, talk, poster**

###**Add title to FacetGrid**
* **assign to g**

###**Add titles to FacetGrid subplots**
