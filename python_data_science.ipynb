#**Data Visualization in Python**

##**Import datasets and libraries**

```python
# Import libraries pandas and numpy
import pandas as pd
import numpy as np

# Import the matplotlib.pyplot submodule and name it plt
import matplotlib.pyplot as plt

# Import datasets from github url
seattle_weather = pd.read_csv('https://raw.githubusercontent.com/nirmalaraj77/Python/main/seattle_weather.csv')
austin_weather = pd.read_csv('https://raw.githubusercontent.com/nirmalaraj77/Python/main/austin_weather.csv')
```

##**METHOD 1**
**figure and subplots**

```python
# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()

# Plot Seattle data, setting data appearance
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], color='b', marker='o',linestyle='--')

# Plot Austin data, setting data appearance
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], color='r', marker='v',linestyle='--')

# Call show to display the resulting plot
plt.show()
```

##**METHOD 2**
**plt.plot**

```python
# Plot Seattle
plt.plot (seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'],color='b', marker='o',linestyle='--')

# Plot Aeattle
plt.plot (austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'],color='r', marker='v',linestyle='--')

# Customize the x-axis label
plt.xlabel ("Time (months)")

# Customize the y-axis label
plt.ylabel ("Precipitation (inches)")

# Add the title
plt.title ("Weather patterns in Austin and Seattle")

# Call show to display the resulting plot
plt.show()
```

```python
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"])

# Customize the x-axis label
ax.set_xlabel ("Time (months)")

# Customize the y-axis label
ax.set_ylabel ("Precipitation (inches)")

# Add the title
ax.set_title ("Weather patterns in Austin and Seattle")

# Display the figure
plt.show()
```
