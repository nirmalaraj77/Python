#**Data Visualization in Python**

##**Use shortcut string notation parameter to specify the marker**
* **fmt**
* **syntax:  marker|line|color**
* **marker|line|color**
* **ms = marker size**
* **mec = marker edge color**
* **mfc = markr face color**
* **ls = linestyle**
* **c = line colour**
* **lw = line width**
* **add multiple lines to same plt**
* **position title loc = left/right/center**

##**Import datasets and libraries**

```python
# Import libraries pandas and numpy
import pandas as pd
import numpy as np

# Import the matplotlib.pyplot submodule and name it plt
import matplotlib.pyplot as plt

# Import datasets from github url
seattle_weather = pd.read_csv('https://raw.githubusercontent.com/nirmalaraj77/Python/main/seattle_weather.csv')
austin_weather = pd.read_csv('https://raw.githubusercontent.com/nirmalaraj77/Python/main/austin_weather.csv')
```

##**Method 1 - figure and subplots**
###**With shortcuts and multiple plots**
* **Create multiple subplots in grid - Small multiples (subplots)**
* **Create figure object with rows and columns of subplots**
* **fig, ax = plt.subplots (r, c)**
* **index into object - use r and c**
* **ax[0,0].plot(x,y)**
###**Special case - two subplots one on top of other**
* **use ax[0], ax[1]**
* **Only 1 xlabel and title**
* **Share y axes range - sharey = True**






```python
# Create multiple Axes with plt.subplots rows and columns
fig, ax = plt.subplots(2,1, sharey=True)

# Plot Seattle data, setting data appearance with shortcuts
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], 'o--b', ms=10)

# Plot Austin data, setting data appearance with shortcuts
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], 'v--r', ms=10)

# Customize the x-axis label
ax[1].set_xlabel("Time (months)")

# Customize the y-axis label
ax[0].set_ylabel ("Precipitation (inches)")
ax[1].set_ylabel ("Precipitation (inches)")

# Add the title
ax[0].set_title ("Weather patterns in Austin and Seattle", loc = 'left')

plt.show()
```

#**Method 2 - Multiple charts with different parameters**
* **Create multiple Axes with plt.subplots**

```python
# PLOT 1

# Create multiple Axes with plt.subplots
fig, ax = plt.subplots()

# Plot Seattle data, setting data appearance with shortcuts
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], 'o--b', ms=10)

# Plot Austin data, setting data appearance with shortcuts
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], 'v--r', ms=10)

# Customize the x-axis label
plt.xlabel ("Time (months)")

# Customize the y-axis label
plt.ylabel ("Precipitation (inches)")

# Add the title
plt.title ("Weather patterns in Austin and Seattle", loc = 'left')


# PLOT 2

# Create multiple Axes with plt.subplots
fig, ax1 = plt.subplots()

# Plot Seattle data, setting data appearance with shortcuts
ax1.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], 'o--b')

# Plot Austin data, setting data appearance with shortcuts
ax1.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], 'o--r')

# Customize the x-axis label
plt.xlabel ("Time (months)")

# Customize the y-axis label
plt.ylabel ("Precipitation (inches)")

# Add the title
plt.title ("Weather patterns in Austin and Seattle")


# Call show to display the resulting plot
plt.show()
```
