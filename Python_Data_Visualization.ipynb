##**Method 1 - figure and subplots**
###**With shortcuts and multiple plots**
* **Create multiple subplots in grid - Small multiples (subplots)**
* **Create figure object with rows and columns of subplots**
* **fig, ax = plt.subplots (r, c)**
* **index into object - use r and c**
* **ax[0,0].plot(x,y)**
###**Special case - two subplots one on top of other**
* **use ax[0], ax[1]**
* **Only 1 xlabel and title**
* **Share y axes range - sharey = True**






```python
# Create multiple Axes with plt.subplots rows and columns
fig, ax = plt.subplots(2,1, sharey=True)

# Plot Seattle data, setting data appearance with shortcuts
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], 'o--b', ms=10)

# Plot Austin data, setting data appearance with shortcuts
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], 'v--r', ms=10)

# Customize the x-axis label
ax[1].set_xlabel("Time (months)")

# Customize the y-axis label
ax[0].set_ylabel ("Precipitation (inches)")
ax[1].set_ylabel ("Precipitation (inches)")

# Add the title
ax[0].set_title ("Weather patterns in Austin and Seattle", loc = 'left')

plt.show()
```

##**twinx Share x axes**
* **ax2 = ax.twinx()**
* **change ticks with ticks_params**

```python
import matplotlib.pyplot as plt

# Initalize a Figure and Axes
fig, ax = plt.subplots()

# Plot the CO2 variable in blue
ax.plot(climate_change.index, climate_change["co2"], color='blue')

# Create a twin Axes that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature in red
ax2.plot(climate_change.index, climate_change["relative_temp"], color='red')

# Change twin axes tick colors with tick_params
ax2.tick_params ('y', colors = 'red')

plt.show()
```

##**Define function**

```python
# Define a function called plot_timeseries
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  # Plot the inputs x,y in the provided color
  axes.plot(x, y, color=color)

  # Set the x-axis label
  axes.set_xlabel(xlabel)

  # Set the y-axis label
  axes.set_ylabel(ylabel, color=color)

  # Set the colors tick params for y-axis
  axes.tick_params('y', colors=color)


# Use the function

fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_change.index, climate_change["co2"], 'blue', "Time (years)", "CO2 levels")

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', "Time (years)", "Relative temperature (Celsius)")

plt.show()
```

##**Annotate**
* **Use ax.annotate (text, xy)**
* **Use pd.Timestamp(x,y)**
* **Place annotate text - xytext =(x,y)**
* **Use arrowprops dict keywords**


```python
fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_change.index, climate_change["co2"], 'blue', "Time (years)", "CO2 levels")

# Create an Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', "Time (years)", "Relative temp (Celsius)")

# Annotate the point with relative temperature >1 degree
ax2.annotate(">1 degree", xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2008-10-06'), -0.2), arrowprops={'arrowstyle':'->', 'color':'gray'})

plt.show()
```

###**Specify hue colors**

* **specify hue colors - palette parameter - dict mapped to colors**

```python
import matplotlib.pyplot as plt
import seaborn as sns


fig, axes = plt.subplots(1,2)
fig.suptitle('Segment data with hue')

# Load built-in dataset
tips = sns.load_dataset('tips')

# Create hue color dictionary for scatter plot
hue_colors_scatter = {'Yes':'black', 'No':'red'}

# Scatter plot with hue and hue order
# Use palette parameter
sns.scatterplot(ax=axes[0],x = 'total_bill', y = 'tip', data = tips, hue = 'smoker',
                hue_order = ['Yes', 'No'], palette = hue_colors_scatter)

# Create hue color dictionary for count plot
hue_colors_count = {'Male':'purple', 'Female':'cyan'}

# Count plot with hue and hue order
# Use palette parameter
sns.countplot(ax=axes[1],x='smoker', data=tips, hue = 'sex',
              hue_order=['Female', 'Male'], palette=hue_colors_count)
plt.show()
```

##**Point plots**
* **mean of a quantitative variable for the observations in each category**
* **plotted as a single point**
* **vertical bars represent the 95% confidence intervals for that mean**
* **confidence intervals show the level of uncertainty about the mean estimates**
* **Assuming data is a random sample of population - can be 95% sure that true population mean in each group lies within the confidence interval shown**
* **line plot has quantitative variable (time) on x-axis**
* **point plot has catergorical variable on x-axis**
* **use kind = 'point'**
* **segregate with hue**
* **disconnect points = join = False**
* **calculate points and ci from median (instead of mean) - import median from numpy and use estimator = median**
* **median more robust to outliers**
* **use ci=None**
* **use capsize = desired width of cap e.x. 0.2**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.get_dataset_names()
diamonds=sns.load_dataset('diamonds')

sns.catplot(x='cut', y='price', data=diamonds, kind='point',
  hue='cut', estimator=np.median,)
```

##**Titles and Labels**

###**FacetGrid vs. AxesSubplot objects**
* **assign the plot output to variable g**
* **type (g)**

* **FacetGrid consists of one or more AxesSubplots**
* **relplot() and catplot() support making subplots**
* **relplot() and catplot() create FacetGrid objects**

###**Add title to FacetGrid**
* **assign to g**
* **use g.fig.suptitle('Title')**
* **set title height - y = 1.03 (default 1)**

###**Add titles to FacetGrid subplots**
* **assign to g**
* **use g.set_titles('Titles')**
* **use g.set_titles ('Titles {col_name})**
* **set title height - y = 1.03 (default 1)**

###**Add title to AxesSubplot**
* **assign to g**
* **use g.set_title('Title')**
* **set title height - y = 1.03 (default 1)**

###**Add axes labels for FacetGrid and AxesSubplot**
* **assign to g**
* **use g.set(xlabel='xlabel', ylabel = 'ylabel')**

###**Rotate x-axis tick label for FacetGrid and AxesSubplot**
* **create plot**

##**Combining plots**
* **create two axes and share y-axis between the plots (sharey=True)**
* **plot full histogram on ax0**
* **use pandas' query to only plot data for Minnesota**
* **new concept - use axvline to denote maximum amount budget for tuition**
* **final step  - turn on and display legend**

* **use nrows and ncols**
* **use figsize**
* **use stat=density**



```python
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(7,4))


sns.histplot(college_data['Tuition'], stat='density', ax=ax0)

sns.histplot(college_data.query('Regions=="South East"')['Tuition'], stat='density', ax=ax1)

ax1.set(xlabel='Tuition (SE)', xlim=(0,70000))
ax1.axvline(x=20000, label='My Budget', linestyle='--')
ax1.legend()

plt.show()
```

```python
# Create a figure and axes. Then plot the data
fig, ax = plt.subplots()
sns.histplot(housing['fmr_1'], ax=ax)

# Customize the labels and limits
ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500), title="US Rent")

# Add vertical lines for the median and mean
ax.axvline(x=housing['fmr_1'].median(), color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=housing['fmr_1'].mean(), color='b', label='Mean', linestyle='-', linewidth=2)

# Show the legend and plot the data
ax.legend()
plt.show()
```

```python
# Create a plot with 1 row and 2 columns that share the y axis label
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True)

# Plot the distribution of 1 bedroom apartments on ax0
sns.histplot(housing['fmr_1'], ax=ax0)
ax0.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 2 bedroom apartments on ax1
sns.histplot(housing['fmr_2'], ax=ax1)
ax1.set(xlabel="2 Bedroom Fair Market Rent", xlim=(100,1500))

# Display the plot
plt.show()
```
